```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
url = "https://comp.wisereport.co.kr/company/c1010001.aspx?cmp_cd="
code = "012450"
```

```python
driver = webdriver.Chrome()
```

```python
# res = requests.get(url+code)
driver.get(url+code)
```

```python
# res 데이터에서 text로 추출한 뒤 파싱
# soup = bs(res.text, 'html.parser')
soup = bs(driver.page_source, 'html.parser')
```

```python
len(
    soup.find_all(
        'table',
        attrs={
            'class' : 'gHead01 all-width'
        }
    )
)
```

```python
# table 태그 중 class 가 gHead01 all-width 인 태그 중 마지막 데이터만 사용
table_data = soup.find_all(
    'table',
    attrs={
        'class' : 'gHead01 all-width'
    }
)[-1]
```

```python
table_data
```

```python
# table_data를 문자열로 변환하고 데이터프레임으로 만든다.
df = pd.read_html(
    str(table_data)
)[0]
```
