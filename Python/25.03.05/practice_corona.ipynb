#### 일일확진자, 일일사망자 컬럼을 추가하는 함수
- 매개변수 2개 생성
    - 첫 인자값 : 데이터프레임
    - 두번째 인자값 : 차순정렬을 할 타겟 컬럼명
- 특정 컬럼을 제거
    - 'Unnamed: 0', 'seq' 컬럼을 제거
- 컬럼의 이름을 변경
    - '등록일시', '총사망자', 총확진자', '기준일', '기준시간', '수정일시', '누적의심자', '누적확진율'
- 특정 컬럼('기준일')의 데이터를 기준으로 오름차순 정렬
- 인덱스를 초기화하고 기존의 인덱스를 제거
- 일일확진자 컬럼을 추가
    - 총 확진자의 차이를 가지고 데이터 생성
- 일일사망자 컬럼을 추가
    - 총 사망자의 차이를 가지고 데이터 생성
- 최종 결과 데이터프레임을 되돌려준다.

```python
def corona_data(_df, _col):
    # 원본의 데이터를 수정하지 않기 위해 복사본 데이터를 생성
    result = _df.copy()
    # 특정 컬럼을 제거
    # Unnamed: 0 컬럼이 존재하면 제거 존재하지 않으면 아무행동도 하지 않는다.
    # 비교 : "Unnamed: 0", 데이터프레임의 컬럼들중 포함되어있는가?
    if 'Unnamed: 0' in result.columns:
        result.drop(["Unnamed: 0", "seq"], axis = 1, inplace = True)
    else:
        result.drop(["seq"], axis = 1, inplace = True)
    # 컬럼의 이름을 변경
    result.columns = ['등록일시', '총사망자', '총확진자', '기준일', '기준시간', '수정일시', '누적의심자', '누적확진율']
    # 특정 컬럼(_col)을 기준으로 오름차순 정렬
    result.sort_values([_col], inplace = True)
    # 인덱스를 초기화하고 기존의 인덱스는 제거
    result.reset_index(drop = True, inplace = True)
    # 특정 컬럼을 생성
    result['일일확진자'] = result['총확진자'].diff().fillna(0)
    result['일일사망자'] = result['총사망자'].diff().fillna(0)
    # 결과를 되돌려준다
    return result
```
