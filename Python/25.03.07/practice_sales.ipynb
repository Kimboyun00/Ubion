```python
import pandas as pd
```

- 연습문제

1. csv폴더 안에 있는 'Sales Records.csv' 파일 로드
2. 'Order ID', 'Total Revenue', 'Total Cost', 'Total Profit', 'Order Reioeity' 컬럼을 제거
3. 컬럼의 이름 변경 : ['대륙', '국가', '상품종류', '판매채널', '주문일자', '발송일자', '판매개수', '단가', '원가']
4. 대륙, 국가를 기준으로 오름차순 정렬
5. 파생변수 생성
    - 총단가 -> 판매개수 * 단가
    - 총원가 -> 판매개수 * 원가
    - 총이윤 -> 총단가 - 총원가

```python
# 파일을 로드
sales = pd.read_csv("../../csv/Sales Records.csv")
```

```python
sales.info()
```

```python
df = sales.copy()
```

```python
# 특정 컬럼을 제거
# drop()
df.drop(
    ["Order ID", "Total Revenue", 'Total Cost', 'Total Profit', 'Order Priority'],
    axis=1,
    inplace=True
)
```

```python
# 컬럼의 이름 변경
df.columns = ['대륙', '국가', '상품종류', '판매채널', '주문일자', '발송일자', '판매개수', '단가', '원가']
```

```python
df.head(1)
```

```python
df.sort_values(
    ['대륙', '국가'],
    inplace=True
    )
df.head()
```

```python
df.reset_index(drop = True, inplace = True)
```

```python
df['총단가'] = df['판매개수'] * df['단가']
df['총원가'] = df['판매개수'] * df['원가']
df['총이윤'] = df['총단가'] - df['총원가']
```

```python
df.head()
```

```python
# 총 이윤의 합계가 가장 높은 국가는 어디인가?
# 국가별로 데이터프레임을 나눠주는 작업
# 그룹화 함수 groupby()
# df[['국가', '총이윤']].groupby('국가').sum()
group_data = df.groupby('국가')[['총이윤']].sum()
```

```python
# 총이윤이 가장 높은 10개의 국가를 출력
group_data.sort_values('총이윤',ascending=False).head(10)
```

```python
# 해당하는 데이터에서 South Korea의 순위는 몇위인가?
# 그룹데이터를 총이윤을 기준으로 내림차순 정렬
group_data.sort_values('총이윤', ascending=False, inplace=True)
```

```python
result = 0
for i in group_data.index:
    # print(i)
    # break
    result += 1
    if i == 'South Korea':
        break
print(result)
```
